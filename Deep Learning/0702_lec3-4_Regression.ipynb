## x,y train data 미리 정의해서 그래프 구축

```python
x_train = [[1.0, 2.0],[1.0,3.0],[2.0,4.0]]
y_train = [5,6,7]

#trainable parameter 정의 및 그래프 구축
#가중치 초기화
tf.set_random_seed(777)
W = tf.Variable(tf.random_normal([2,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1], name = 'bias'))

hypothesis = tf.matmul(x_train, W) + b
tf.print(hypothesis)
```

```python
#어떻게 train할지 정해야됨 
#cost, optimizer
cost = tf.reduce_mean(tf.square(hypothesis - y_train)) #costfunction 정의
opt = tf.train.GradientDescentOptimizer(learning_rate= 0.01) #optimizer 정의
train = opt.minimize(cost) #cost가 최소화되는 방향으로 최적화
```

```python
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
      print(step, sess.run(cost), sess.run(cost), sess.run(b))
```

## Placeholder 이용

1. 세션 생성
2. tf.global_variables_initializer() run해주기
3. sess.run([cost,W,b,train])



```python
x = tf.placeholder(tf.float32, name='x', shape = [None, 2])
y = tf.placeholder(tf.float32, name='y')

#trainable parameter 정의 및 그래프 구축
#가중치 초기화
tf.set_random_seed(777)
W = tf.Variable(tf.random_normal([2,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

# Hypothesis

hypothesis = tf.matmul(x, W) + b


#어떻게 train할지 정해야됨 
#cost, optimizer
cost = tf.reduce_mean(tf.square(hypothesis - y)) #costfunction 정의
opt = tf.train.GradientDescentOptimizer(learning_rate= 0.01) #optimizer 정의
train = opt.minimize(cost) #cost가 최소화되는 방향으로 최적화

```

```python
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(201):
    cost_val, W_val, b_val, _ = \
      sess.run([cost, W, b, train], feed_dict = {x: [[1,2],[1,3],[2,4]], y: [4, 5, 6]})

    if step % 20 == 0:
      print(step, cost_val, W_val, b_val)
  print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ\n")
  print(sess.run(hypothesis, feed_dict={x: [[1, 3]]}))
```
